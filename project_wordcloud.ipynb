```python
result=[]
end_point='https://openapi.naver.com/v1/search/news.json?start='
start=1
while(start<=1000):
    url=end_point+str(start)
    data=json.loads(requests.get(url,
                params=parameter, headers=myheader).text)
    result=result+data['items']
    start=start+parameter['display']
```
