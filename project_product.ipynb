```python
end_point='https://search.shopping.naver.com/search/all?frm=NVSHMDL&origQuery=%ED%9C%A0%EC%B2%B4%EC%96%B4&pagingSize=40&productSet=model&query=%ED%9C%A0%EC%B2%B4%EC%96%B4&sort=rel&timestamp=&viewType=list&pagingIndex='
```

```python
name=[]
price=[]
for page in range(1, 10):
    url=end_point+str(page)
    driver.get(url)
    
    tags_name=driver.find_elements_by_css_selector('#__next > div > div.style_container__1YjHN > div.style_inner__18zZX > div.style_content_wrap__1PzEo > div.style_content__2T20F > ul > div > div > li > div > div.basicList_info_area__17Xyo > div.basicList_title__3P9Q7 > a')
    for tag in tags_name:
        name.append(tag.text)
        
    tags_price=driver.find_elements_by_css_selector('#__next > div > div.style_container__1YjHN > div.style_inner__18zZX > div.style_content_wrap__1PzEo > div.style_content__2T20F > ul > div > div > li > div > div.basicList_info_area__17Xyo > div.basicList_price_area__1UXXR > strong > span > span.price_num__2WUXn')
    for tag in tags_price:
        price.append(tag.text)

    
```

```python
name
```

```python
price
```

```python
df=pd.DataFrame({'제품':name, '가격':price})
df
```
