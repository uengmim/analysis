```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import requests
import xmltodict
import json
#E나라지표
```

```python
df_type=pd.read_csv('data/등록_장애인수1.csv')
df_type
```

```python
df_type1=df_type.groupby('장애유형별')[['2020']].sum()
df_type1
```

```python
plt.figure(figsize=(16,9))
plt.plot(df_type1['2020'], marker='o', label='2020')
plt.xlabel('장애유형별', size = 25)
plt.ylabel('장애인 수', size = 25)
plt.title('유형별 장애인 인구 수 증가', size = 30)
plt.grid(alpha=1)
plt.xticks(rotation=50, size=20)
plt.legend(fontsize=20)
plt.show()
```

```python
df_year=pd.read_csv('data/등록_장애인수.csv')
df_year
```

```python
df_year1=df_year.groupby('연도별')[['합계','지체','청각ㆍ언어','뇌병변']].sum()
df_year1
```

```python
plt.figure(figsize=(16,9))
plt.plot(df_year1['합계'], marker='^', label='합계')
plt.plot(df_year1['지체'], marker='o', label='지체')
plt.plot(df_year1['청각ㆍ언어'], marker='o', label='청각ㆍ언어')
plt.plot(df_year1['뇌병변'], marker='o', label='뇌병변')
plt.xlabel('연도별', size = 25)
plt.ylabel('장애인 수', size = 25)
plt.title('연도별 장애인 인구 수 증가', size = 30)
plt.grid(alpha=1)
plt.xticks(rotation=50, size=20)
plt.legend(fontsize=20)
plt.show()
```
